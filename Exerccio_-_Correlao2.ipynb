### Correlação entre Pontos e gols marcados

```python
import pandas as pd
import numpy as np

df = pd.read_csv("cbf_2020.csv")
gols_marcados = df[['Pontos', 'Gols Marcados']].corr()
gols_marcados
```

### Correlação entre Pontos e gols sofridos

```python
gols_sofridos = df[['Pontos', 'Gols Sofridos']].corr()
gols_sofridos
```

### Análise dos Resultados

Qual o nível (fraco, moderado, forte) das correlações entre as variáveis analisadas

A correlação entre gols marcados e a pontuação é "Muito Forte", visto que os pontos crescem a medida que os pontos crescem tão tendencialmente quanto aos gols marcados.

Já a correlação entre gols sofridos e a pontuação é "Forte", mas a correlação é anticorrelativa, visto que os pontos diminuem a medida que os gols sofridos aumentam. Porém, quanto menos gols forem marcados, a correlação entre os pontos serem baixos é "Forte".
