## Pandas

A partir do dataset `movies.csv` utilize o Pandas para responder as seguintes perguntas:

### Qual o gênero de filme mais lucrativo em todo o mundo?

```python
import pandas as pd

filmes = pd.read_csv("movies.csv")

filmes["Lucro em %"] = 100*((filmes["Worldwide Gross"] - filmes["Production Budget"]) / filmes["Production Budget"])

filmes.groupby(["Major Genre"]).mean().sort_values(by="Lucro em %", ascending = False)["Lucro em %"]

#Documentários tiveram a maior percentual de lucro em relação ao gasto na produção.
```

### Qual o diretor dirigiu mais filmes?

```python
from collections import Counter
contagem = Counter(filmes['Director']).most_common(5)
contagem

#Vemos que o que dirigiu mais filmes foi o NaN... Brincadeira, o NaN é quando o diretor não foi informado. 
#Seguido dele, está o Steven Spielberg, que ele sim foi que efetivamente mais dirigiu filmes na lista.'''
```

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
top_5 = filmes[['Title', 'Lucro em %']].sort_values(by='Lucro em %', ascending=False).head(5)
top_5
#A lista dos 5 filmes que mais arrecadaram grana abaixo! Mas tá muito complicado de entender esse percentual.
```

```python
#Vamo dar uma arrumada nisso, só pra dar uma visualizada melhor.

for i in range(top_5.shape[0]):
    numero = top_5['Lucro em %'].iloc[i]
    top_5['Lucro em %'].iloc[i] = f'{numero:.2f}'
top_5

#Pronto, agora podemos ver que esses valores referem ao percentual de lucro. Além disso, o fato de que o filme Atividade Paranormal
#Rendeu mais de 1.000.000% do capital investido é impressionante.
```
