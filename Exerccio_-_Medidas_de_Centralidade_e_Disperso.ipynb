## Medidas de Centralidade e Dispersão

A partir do dataset `movies.csv` calcule a média, mediana e desvio padrão para as colunas:

### US Gross

```python
import pandas as pd
import numpy as np

'''def mediana (x):
    
    serie = x[x >= 0].sort_values()
    mediana = None
    meio = int(len(serie)/2)
    
    if len(serie) % 2 == 0:
        mediana = (serie.iloc[meio] + serie.iloc[meio - 1])/2
        
    else:
        meio = int(meio + 0.5)
        mediana = serie.iloc[meio]
        
    return mediana'''
        
def mostrar_dados (x):
    
    print(f'Média = {x.mean():.2f}, Mediana = {x.median()}, Desvio padrão = {x.std():.2f}')
          
    return

filmes = pd.read_csv("movies.csv")
us_gross = filmes["US Gross"]


mostrar_dados(us_gross)
```

```python
us_gross.median()
```

### Production Budget

```python
production_budget = filmes["Production Budget"]
mostrar_dados(production_budget)
```

### Running Time

```python
running_time = filmes["Running Time (min)"]
mostrar_dados(running_time)
```

### IMDB Rating

```python
imdb_rating = filmes["IMDB Rating"]
mostrar_dados(imdb_rating)
```

### IMDB Votes

```python
imdb_votes = filmes["IMDB Votes"]
mostrar_dados(imdb_votes)
```
